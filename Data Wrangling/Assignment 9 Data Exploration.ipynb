```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
from sklearn.datasets import load_diabetes

# Load the dataset
diabetes = load_diabetes()
data = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
data['target'] = diabetes.target
```

```python
data.head().round(2)
```

```python
data.describe().round(2)
```

```python
# ARGS and Keyword Args
# Create a figure of figsize = (10,6)
# (Hint: plt.figure(figsize=(10,6))
# Plot the first feature ('age') against the 'target' variable using positional arguments. Include a marker, line style, and color keywords.
# Add a title for your plot and labels for x and y axis.
plt.figure(figsize=(10,6))
plt.scatter(data['age'], data['target'], marker='^', linestyle='--', color='g')
plt.xlabel('Age')
plt.ylabel('Target')
plt.title('Age vs. Target')
plt.show()
```

```python
# Figures and Subplots
# Create a figure with two subplots and figsize=(14,6)
# (Hint: fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
# Use a scatter plot for subplot 1 and plot 'bmi' vs 'target'. Include color and marker keywords.
# Add title and labels for x and y axis to the subplot using set_title, set_xlabel, and set_ylabel.
# Use a scatter plot for subplot 2 and plot ' 'bp' vs 'target'. Include different color and different marker from that of subplot 1.
# Add title and labels for x and y axis to the subplot using set_title, set_xlabel, and set_ylabel.
# Display the plot.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
ax1.scatter(data['bmi'], data['target'], marker='s', color='r')
ax1.set_xlabel('BMI')
ax1.set_ylabel('Target')
ax1.set_title('BMI vs. Target')

ax2.scatter(data['bp'], data['target'], marker='o', color='b')
ax2.set_xlabel('BP')
ax2.set_ylabel('Target')
ax2.set_title('BP vs. Target')
plt.show()
```

```python
# Ticks, Labels, and Legends
# Create a figure of figsize = (10,6)
# Use a scatter plot and customize the plot for 's1' vs 'target'. Include color, marker and label keywords.
# Add ticks with fontsize of 10, and labels for x and y axis to the plot.
# Add legend and grid to the plot.
# Display the plot.
plt.figure(figsize=(10,6))
plt.scatter(data['s1'], data['target'], marker='D', color='g', label='S1 vs Target')
plt.xlabel('S1', fontsize=10)
plt.ylabel('Target', fontsize=10)
plt.title('S1 vs. Target')
plt.tick_params(axis='both', labelsize=10)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Colors, Markers, and Line Styles
# Create a figure of figsize = (10,6)
# Plot 's1', 's2', and 's3' each against 'target' using different colors, markers, and line styles.
# Add title and labels for x and y axis to the plot.
# Add legend to the plot.
# Display the plot.
plt.figure(figsize=(10,6))

plt.subplot(1, 3, 1)
plt.scatter(data['s1'], data['target'], marker='s', color='r', linestyle='--', label='S1 vs Target')
plt.xlabel('S1', fontsize=10)
plt.ylabel('Target', fontsize=10)
plt.title('S1 vs. Target')
plt.tick_params(axis='both', labelsize=10)
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 2)
plt.scatter(data['s2'], data['target'], marker='^', color='g', linestyle='--', label='S2 vs Target')
plt.xlabel('S2', fontsize=10)
plt.ylabel('Target', fontsize=10)
plt.title('S2 vs. Target')
plt.tick_params(axis='both', labelsize=10)
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 3)
plt.scatter(data['s3'], data['target'], marker='o', color='b', linestyle='--', label='S3 vs Target')
plt.xlabel('S3', fontsize=10)
plt.ylabel('Target', fontsize=10)
plt.title('S3 vs. Target')
plt.tick_params(axis='both', labelsize=10)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Legends and grid
# Create a figure of figsize = (10,6).
# Plot 's4' vs 'target' and add legends and grids to the plot.
# Add grid and legend with loc =2 to the plot.
# Display the plot.
plt.figure(figsize=(10,6))
plt.scatter(data['s4'], data['target'], marker='D', color='g', label='S4 vs Target')
plt.xlabel('S4', fontsize=10)
plt.ylabel('Target', fontsize=10)
plt.title('S4 vs. Target')
plt.tick_params(axis='both', labelsize=10)
plt.legend(loc=2)
plt.grid(True)
plt.show()
```

```python
# Saving Plots as an Image File
# Create a figure of figsize = (10,6).
# Plot  'age' vs 'target' and add a marker.
# Add title and labels for x and y axis to the plot.
# Save the plot of 'age' vs 'target' as both PNG and PDF files.
# Display the plot
plt.figure(figsize=(10,6))
plt.scatter(data['age'], data['target'], marker='^', color='b', label='Age vs Target')
plt.xlabel('Age', fontsize=10)
plt.ylabel('Target', fontsize=10)
plt.title('Age vs. Target')
plt.tick_params(axis='both', labelsize=10)
plt.legend(loc=2)
plt.grid(True)
plt.savefig('Age vs. Target.png')
plt.savefig('Age vs. Target.pdf')
plt.show()
```
