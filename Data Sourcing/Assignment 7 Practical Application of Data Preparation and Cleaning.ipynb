Handle missing values

```python
check_null_values = fifa.isnull().sum()
show_null_values = check_null_values[check_null_values > 0]  # Show only columns with null values
print(show_null_values)
```

```python
fifa.columns
```

```python
# Replace the missing values with the respective mean in all columns
for i in ['Attacking', 'Crossing', 'Finishing',
       'Heading Accuracy', 'Short Passing', 'Volleys', 'Skill', 'Dribbling',
       'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control', 'Movement',
       'Acceleration', 'Sprint Speed', 'Agility', 'Reactions', 'Balance',
       'Power', 'Shot Power', 'Jumping', 'Stamina', 'Strength', 'Long Shots',
       'Mentality', 'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Penalties', 'Composure', 'Defending', 'Marking', 'Standing Tackle',
       'Sliding Tackle', 'Goalkeeping', 'GK Diving']:
    fifa[i].fillna(fifa[i].mean(), inplace=True)

# Check for remaining missing values if any
print(fifa.isnull().sum().any())
```

Remove duplicates

```python
# Check for duplicates
fifa.duplicated().sum()
```

```python
# Drop duplicates
fifa.drop_duplicates(inplace=True)
fifa.duplicated().sum()
```

Standardize data formats & Correct data types

```python
fifa[['Joined','Height','Weight','Value','Wage','Release Clause']].head()
```

```python
fifa[['Joined','Height','Weight','Value','Wage','Release Clause']].info()
```

```python
# Joined: convert to datetime dtype
fifa['Joined'] = pd.to_datetime(fifa['Joined'], format='%d-%b-%y')
# Feature engineering: add new columns Year & Month
fifa['Year'] = fifa['Joined'].dt.year
fifa['Month'] = fifa['Joined'].dt.month
```

```python
# Weight: convert from lbs to kg and change to float dtype
fifa['Weight'] = fifa['Weight'].str.replace("lbs", "").astype(float).apply(lambda x: round(x / 2.2, 1))
# Height: convert from feet to cm and change to float dtype
fifa['Height'] = fifa['Height'].apply(lambda x: round(float(x.split("'")[0]) * 30.48 + float(x.split("'")[1].replace('"', '')) * 2.54, 1))
```

```python
# Value, Wage & Release Clause: remove symbols and change to float dtype
def clean_value(amount):
    if 'M' in amount:
        return float(amount.replace('â‚¬', '').replace('M', '')) * 1000000
    elif 'K' in amount:
        return float(amount.replace('â‚¬', '').replace('K', '')) * 1000
    else:
        return float(amount)

fifa['Value'] = fifa['Value'].apply(clean_value)
fifa['Wage'] = fifa['Wage'].apply(clean_value)
fifa['Release Clause'] = fifa['Release Clause'].apply(clean_value)
```
